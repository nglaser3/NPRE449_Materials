```python
import numpy as np
import scipy.linalg as sla
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,1,1000)
v = lambda r: 1- r**2
t = lambda r: r**4/ 3- 4/3 * r**2 + 1
```

```python
plt.plot(x,v(x),label = 'Normalized Velocity')
plt.plot(x,t(x), label = 'Normalized Temperature')
plt.legend()
plt.grid()
plt.savefig('plots/q2.png',dpi=600)
plt.show()
```
